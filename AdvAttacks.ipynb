# Analysis of files and directories



*   **ToS/**

Main directory of the dataset

*   **ToS/Labels** 

Contains the files with tags per line. Values are {-1, +1}

*   **ToS/Labels_A**

Type of clause: Arbitration

*   **ToS/Labels_CH**

Type of clause: Unilateral change

*   **ToS/Labels_CR**

Type of clause: Content removal

*   **ToS/Labels_J**

Type of clause: Jurisdiction

*   **ToS/Labels_LAW**

Type of clause: Choice of law

*   **ToS/Labels_LTD**

Type of clause: Limitation of liability

*   **ToS/Labels_TER**

Type of clause: Unilateral termination

*   **ToS/Labels_USE**

Type of clause: Contract by using

*   **OriginalTaggedDocuments**

Contains the original ToS sentences and includes tags of the classification. For example:

> 21.2 \<law2>If you reside in China these Terms will be interpreted in accordance with the laws of China (“China Laws”).\</law2> \<a2>Any dispute arising from or in connection with this Agreement shall be submitted to the China International Economic and Trade Arbitration Commission (CIETAC) for arbitration in Beijing which shall be conducted in accordance with CIETAC’s arbitration rules in effect at the time of applying for arbitration, provided that this section shall not be construed to limit any rights which Airbnb may have to apply to any court of competent jurisdiction for an order requiring you to perform or be prohibited from performing certain acts and other provisional relief permitted under China Laws or any other laws that may apply to you.\</a2> The arbitration proceedings shall be conducted in English. The arbitral award rendered is final and binding upon both parties.

> 21.3 \<law2>If you reside outside of the United States and China, these Terms will be interpreted in accordance with Irish law.\</law2> The application of the United Nations Convention on Contracts for the International Sale of Goods (CISG) is excluded. The choice of law does not impact your rights as a consumer according to the consumer protection regulations of your country of residence. \<j3>If you are acting as a consumer, you agree to submit to the non-exclusive jurisdiction of the Irish courts.\</j3> \<j1>Judicial proceedings that you are able to bring against us arising from or in connection with these Terms may only be brought in a court located in Ireland or a court with jurisdiction in your place of residence.\</j1> \<j1>If Airbnb wishes to enforce any of its rights against you as a consumer, we may do so only in the courts of the jurisdiction in which you are a resident.\</j1> If you are acting as a business, you agree to submit to the exclusive jurisdiction of the Irish courts.

*   **Postags**

Contains tags per line of how each word has been interpreted. For example (Academia line 28):

> PRP RB VBP TO PRP$ NN TO CC NN IN JJ JJ NN IN PRP . 

*   **Sentences**

Contains the lower case of each sentence, such as (Academia line 28):

> you hereby consent to our access to and collection of such personal information about you . 

*   **Trees**

Contains a tree structure of each sentence, word and tag. The following is an instance of Academia line 28:

> (ROOT (S (NP (PRP You)) (ADVP (RB hereby)) (VP (VP (VBP consent) (PP (TO to) (NP (PRP$ our) (NN access) (S (VP (TO to)))))) (CC and) (VP (NP (NP (NN collection)) (PP (IN of) (NP (JJ such) (JJ personal) (NN information)))) (PP (IN about) (NP (PRP you))))) (. .)))

### **TFIDF**

```python
# All sentences of all documents are gathered

sentences_all_docs = []

for i in obj_sentences_per_file.keys():
  sentences_all_docs.extend(obj_sentences_per_file[i]["sentences"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import random
import pandas as pd
```

```python
tfidf_vectorizer = TfidfVectorizer(min_df=5)#TfidfVectorizer(tokenizer=sentences_custom.get_tokens_spacy, min_df=5, ngram_range=(1,1))
tfidf_vectorizer = tfidf_vectorizer.fit(sentences_all_docs)
tfidf_features_skl = tfidf_vectorizer.get_feature_names_out()
```

```python
def top_tfidf_features(row, features, top_n=15):
    ''' Get top n tfidf values in row and return them with their corresponding feature names.'''
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids]
    df = pd.DataFrame(top_feats)
    df.columns = ['feature', 'tfidf']
    return df


def top_features_in_doc(Xtr, features, row_id, top_n=15):
    ''' Top tfidf features in specific document (matrix row) '''
    xtr_row = Xtr[row_id]
    if type(xtr_row) is not np.ndarray:
        xtr_row = xtr_row.toarray()
    row = np.squeeze(xtr_row)
    return top_tfidf_features(row, features, top_n)


def span_top_tfidf(spans_txt, spans_tfidf, features, index):
    print('span text:\n'+spans_txt[index]+'\n')
    print(top_features_in_doc(spans_tfidf, features, index))
```
