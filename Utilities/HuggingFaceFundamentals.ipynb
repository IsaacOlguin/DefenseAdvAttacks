```python
from transformers import pipeline
```

#### **Sentiment analysis**

```python
classifier = pipeline("sentiment-analysis") #No model was supplied, defaulted to distilbert-base-uncased-finetuned-sst-2-english
classifier("We're very happy to show everyone the Transformers' library")

# [{'label': 'POSITIVE', 'score': 0.999777615070343}]
```

#### Examples of positive and negative sentiment analysis

```python
results = classifier([
                      "They're happy to show all of you the transformers' library",
                      "We hope you don't hate it",
                      "Many people hate to learn things by heart",
                      "It's unbelievable the level of insecurity"
])

for res in results:
  print(f'Label: {res["label"]}, with score: {round(res["score"], 4)}')
```

#### **Speech recognizer**

```python
import torch
from transformers import pipeline

speech_recognizer = pipeline("automatic-speech-recognition", model="facebook/wav2vec2-base-960h")
```

```python
!pip install datasets
```

```python
from datasets import load_dataset, Audio

dataset = load_dataset("PolyAI/minds14", name="en-US", split="train")
```

```python
dataset = dataset.cast_column("audio", Audio(sampling_rate=speech_recognizer.feature_extractor.sampling_rate))
```

```python
result = speech_recognizer(dataset[:4]["audio"])
print([d["text"] for d in result])
```

#### **Using model and tokenizer**

```python
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

Then, it's possible to specify the model and tokenizer in the pipeline() and apply the classifier on the target text.

```python
classifier = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
classifier("Nous sommes très heureux de vous présenter la bibliothèque Transformers")
```
